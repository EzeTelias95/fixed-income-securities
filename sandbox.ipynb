```python
from models.FixedIncomeInstruments import Bullet
bullet = Bullet(maturity_date='2040-10-01',issue_date='2020-10-01', principal=1000, interest_rate=0.05, annual_compound=2)
print('Cashflows', bullet.Cashflows())

total = 0
for cashflow in bullet.Cashflows():
    total += cashflow[1] 
print('Cashflow SUM', total)
print('Present Value Interest Factor of Annuity', bullet.PVIFA(bullet.TotalPeriods()))
print('Coupon Present Value of Ordinary Annuity', bullet.PresentValueOfCoupon_Annuity())
print('Present value of Par: ', bullet.PresentValueOfPar())
print('Bond Price: ', bullet.Price())
```
